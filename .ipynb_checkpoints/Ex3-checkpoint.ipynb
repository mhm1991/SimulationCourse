```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
from math import sqrt
from random import seed
from random import random
from time import time

seed(time)
number_of_time_lapses = 10**2

duration = 10**2
sample_time = []
sample_seq = []
pivot = 0
while pivot < duration:
    #if pivot < 10**(0):
    sample_time.append(10**(-2))
    pivot += 10**(-3)
    sample_seq.append(pivot)
#    elif pivot >= 1:
#        sample_time.append(10**(-2))
#        pivot += 10**(-2)
#        sample_seq.append(pivot)

number_of_time_lapses = len(sample_time)
print(number_of_time_lapses)





plt.rc('text', usetex=True)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

mu, sigma2_theta ,sigma2_x, sigma2_y = 0,1,1,1  # mean and standard deviation
Iteration_number = 1
avg_dist = 0
W_theta = np.random.normal(mu, 1, number_of_time_lapses)
W_x = np.random.normal(mu, 1, number_of_time_lapses)
W_y = np.random.normal(mu, 1, number_of_time_lapses)
theta = list(range(number_of_time_lapses))
x = list(range(number_of_time_lapses))
y = list(range(number_of_time_lapses))
r = list(range(number_of_time_lapses))
theta[0] = 0
x[0] = 0
y[0] = 0
j = 0
colors = ["red","green","orange","black","blue"]
for Pe in [500]:#,10,50,200,500]:
    for i in range(1,number_of_time_lapses):
        time_lapse = sample_time[i]
        theta[i] = theta[i-1] + W_theta[i-1]*sqrt(2*time_lapse)#*(math.pi/180)


    for i in range(1,number_of_time_lapses):
        time_lapse = sample_time[i]
        x[i] = x[i-1] + Pe*math.cos(theta[i-1])*time_lapse + W_x[i-1]*sqrt(2*time_lapse)
        y[i] = y[i-1] + Pe*math.sin(theta[i-1])*time_lapse + W_y[i-1]*sqrt(2*time_lapse)
        r[i] = sqrt(x[i]**2 + y[i] ** 2)
    

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1],labels[::-1],title ='Peclet Numbers',loc = 'upper left')
cord = np.empty((2,number_of_time_lapses))
cord[0,:] = x
cord[1,:] = y
#avg_dist = dist_tau/avg_count
#print(avg_dist)
```
